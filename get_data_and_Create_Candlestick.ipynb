```python
import yfinance as yf
import mplfinance as mpf

# Step 1: Download BTC Data from Yahoo Finance
btc = yf.Ticker("BTC-USD")
btc_data = btc.history(period="max")

# Step 2: Create Candlestick Chart with Customization
mpf.plot(btc_data, 
         type='candle', 
         style='yahoo', 
         title='BTC Candlestick Chart',
         figsize=(12, 8),   # Adjust the figure size
         show_nontrading=True,   # Show non-trading days
         tight_layout=True,   # Automatically adjust subplot parameters for better layout
         scale_width_adjustment=dict(candle=1, volume=1.5),   # Adjust the width of candlesticks and volume bars
         mav=(50, 200),   # Moving averages (adjust as needed)
         addplot=[
             mpf.make_addplot(btc_data['Close'], panel=1, color='b', secondary_y=False)   # Add another plot, e.g., Close prices
         ],
         returnfig=True   # Return the figure object for additional customization
         )

# Additional customization of the figure (if needed)
# fig, axlist = mpf.plot(...) returns the figure object and the list of axes
# You can further customize the figure using matplotlib functions

```
